```python
import random
import matplotlib.pyplot as plt
```

```python
class RandomWalk:
    def __init__(self, number_of_points=100000):
        self.number_of_points = number_of_points
        self.x_value = []
        self.y_value = []
        self.x = random.randint(-200, 200)
        self.y = random.randint(-200, 200)

    def fill(self):
        while len(self.x_value) < self.number_of_points:
            self.x += random.choice([-1, 1]) * random.choice([0, 1, 2, 3, 4])
            self.y += random.choice([-1, 1]) * random.choice([0, 1, 2, 3, 4])
            if self.x == 0 and self.y == 0:
                continue
            self.x_value.append(self.x)
            self.y_value.append(self.y)

```

```python
rw = RandomWalk()
rw.fill()

fig = plt.figure(figsize=(10, 6), dpi=600)
point_color = list(range(rw.number_of_points))   # 1 - 100 000
plt.scatter(rw.x_value, rw.y_value, c=point_color,
            cmap=plt.cm.hsv, edgecolor="none", s=5)


plt.axis('off')
```
