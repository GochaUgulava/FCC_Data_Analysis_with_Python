```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime as dt

filename = "sitka_weather_2014.csv"
place = 'Sitka, AL'
```

```python
with open(filename, 'r') as f1:
    reader = csv.reader(f1)
    header_row = next(reader)
    for counter, column_header in enumerate(header_row):
        print(counter, column_header)
    dates, highs, lows = [], [], []
    for row in reader:
        try:
            date = dt.strptime(row[0], '%Y-%m-%d')
            high = round(5 / 9 * (int(row[1])-32), 1)
            low = round(5 / 9 * (int(row[3])-32), 1)
        except ValueError:
            print('missing value')
        else:
            dates.append(date)
            highs.append(high)
            lows.append(low)
```

```python
fig = plt.figure(figsize=(15, 8))
plt.plot(dates, highs, c='red')
plt.plot(dates, lows, c='blue')
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.2)
plt.title('Daily high and low temperatures - 2014 ' + place, fontsize=18)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel('Temperature, C', fontsize=12)
plt.tick_params(axis='both',  which='major', labelsize=12)
```
